```python
# create density plot for dropout
plt.figure(figsize=(10, 6))
sns.kdeplot(data['dropout'], shade=True, color='gray')
plt.title('Density Plot of Dropout')
plt.show()
```

```python
# value counts for dropout
data['dropout'].value_counts(), data['dropout'].value_counts(normalize=True)
```

```python
# create density plot for cows_predose and cows_postdose
plt.figure(figsize=(10, 6))
sns.kdeplot(data['cows_predose'], label='cows_predose', color='blue', shade=True)
sns.kdeplot(data['cows_postdose'], label='cows_postdose', color='gray', shade=True)
plt.xlabel('COWS')
plt.ylabel('Density')
plt.title('COWS Predose vs Postdose')
plt.legend();
```
